```python
class Node:
    def __init__(self,val):
        self.val = val
        self.next = None
```

```python
class Ring:
    def __init__(self):
        self.head = None
```

```python
def __str__(self):
    if self.head is None:
        return
    ret_str = '['
    temp = self.head
    while temp is not None:
        ret_str += str(temp.val) + ', '
        temp = temp.next
        
        if temp == self.head:
            break
    ret_str = ret_str.rstrip(', ')
    ret_str += ']'
    
    return ret_str

Ring.__str__ = __str__
```

```python
def get_last(self):
    if self.head is None:
        return
    
    temp = self.head.next
    
    while temp.next != self.head:
        temp = temp.next
        
    return temp

Ring.get_last = get_last
```

```python
def length(self):
    if self.head is None:
        return
    
    temp = self.head
    count = 0
    while temp is not None:
        temp = temp.next
        count += 1
        
        if temp == self.head:
            break
    return count

Ring.length = length
```

```python
def get(self, index):
    if index == 0:
        return self.head.val
    
    temp = self.head
    count = 0
    while temp is not None and count < index:
        temp = temp.next
        count += 1
        
        if temp == self.head:
            break
            
    if count != index:
        raise Exception('No value found on this index.')
        
    return temp.val

Ring.get = get
```

```python
def push(self, val):
    new_node = Node(val)
    
    if self.head is None:
        self.head = new_node
        new_node.next = self.head
        return
    
    last = self.get_last()
    
    last.next = new_node
    new_node.next = self.head
    
    
Ring.push = push
```

```python
def pop(self):
    if self.head is None:
        return
    
    last = self.get_last()
    
    if last == self.head:
        self.head = None
    
    temp = self.head
    
    while temp != last:
        prev = temp
        temp = temp.next
        
    last = prev
    prev.next = self.head

Ring.pop = pop    
```

```python
def insert(self, index, val):
    new_node = Node(val)
    
    last = self.get_last()
    
    if self.head is None:
        self.head = new_node
        self.head.next = self.head
        return
    
    if index == 0:
        new_node.next = self.head
        self.head = new_node
        last.next = self.head
        return
    
    temp = self.head
    counter = 0
    
    while counter < index:
        prev = temp
        temp = temp.next
        counter += 1
        
        #if temp == last:
         #   prev = temp
          #  temp = self.head
           # counter += 1
            #continue
    
    new_node.next = temp
    prev.next = new_node
    
    if new_node.next == self.head:
        self.head = new_node
        
Ring.insert = insert
```

```python
def remove_val(self, val):
    if self.head is None:
        raise Exception("List is already empty")
        
    temp = self.head
    last = self.get_last()

    if temp.val == val:
        if last == self.head:
            self.head = None
        else:
            self.head = temp.next
            last.next = self.head
        return
    
    prev = temp
    temp = temp.next
    while temp != self.head:
        if temp.val == val:
            break
        prev = temp
        temp = temp.next
        
    if temp == self.head:
        raise Exception("Value not found")
    
    prev.next = temp.next
    
    
Ring.remove_val = remove_val
```

```python
def remove_index(self,index):
    if self.head is None:
        raise Exception("List is empty")
    
    last = self.get_last()
    
    if index == 0:
        if last == self.head:
            self.head = None
            return
        else:
            self.head = self.head.next
            last.next = self.head
        return
    
    temp = self.head
    prev = None
    count = 0
    
    while count < index:
        prev = temp
        temp = temp.next
        count += 1
    
    if temp == self.head:
        prev.next = temp.next
        self.head = temp.next
        
    prev.next = temp.next

    if self.head is None:
        raise Exception("List is empty")
        
Ring.remove_index = remove_index        
```
